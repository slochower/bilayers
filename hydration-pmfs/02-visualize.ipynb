```r
width = 5 * 1.2 * 1.2
height = 3.5 * 2
options(repr.plot.width=width, repr.plot.height=height)


mg = file.path('mg-water-frame-3052-all-2.28.csv') %>%
  readr::read_csv(col_names=c('distance', 'mean', 'std'))

mg_pmf = file.path('../../PMF-45/magnesium.dat') %>%
  readr::read_table(col_names=c('distance', 'pmf'))

ca = file.path('ca-water-all-2.60.csv') %>%
  readr::read_csv(col_names=c('distance', 'mean', 'std'))

ca_pmf = file.path('../../PMF-45/calcium-2.dat') %>%
  readr::read_table(col_names=c('distance', 'pmf'))

colors <- brewer.pal(8,"Set1")


gg_ca = ca %>%
    dplyr::filter(distance < 6) %>%
    dplyr::group_by(distance) %>%
    dplyr::summarize(mean_waters = mean(mean)) %>%
    ggplot2::ggplot(ggplot2::aes(x = distance, y = mean_waters)) +
    ggplot2::geom_smooth() + 
    ggplot2::geom_jitter() + 
    ggplot2::xlab('') + 
    ggplot2::ylab('') + 
    ggplot2::ggtitle(expression(paste("Waters within 2.28 Å of Ca"^{textstyle("2+")}))) + 
    ggplot2::theme_bw() +
    ggplot2::scale_x_continuous(breaks=seq(0, 6, 1), limits=c(1, 6)) + 
    ggplot2::scale_y_continuous(breaks=seq(3, 7, 1), limits=c(3, 7)) + 
    ggplot2::theme(legend.position="none",  plot.margin = ggplot2::unit(c(0,0,0,0), "cm")) +
    ggplot2::scale_fill_manual(name = NULL, values = c('#377eb8')) +
    ggplot2::scale_color_manual(values=c('#377eb8'))

gg_mg = mg %>%
    dplyr::filter(distance < 6) %>%
    dplyr::group_by(distance) %>%
    dplyr::summarize(mean_waters = mean(mean)) %>%
    ggplot2::ggplot(ggplot2::aes(x = distance, y = mean_waters)) +
    ggplot2::geom_smooth() + 
    ggplot2::geom_jitter() +
    ggplot2::xlab('') + 
    ggplot2::ylab('') + 
    ggplot2::ggtitle(expression(paste("Waters within 2.60 Å of Mg"^{textstyle("2+")}))) + 
    ggplot2::theme_bw() +
    ggplot2::scale_x_continuous(breaks=seq(0, 6, 1), limits=c(1, 6)) + 
    ggplot2::scale_y_continuous(breaks=seq(3, 7, 1), limits=c(3, 7)) + 
    ggplot2::theme(legend.position="none", plot.margin = ggplot2::unit(c(0,0,0,0), "cm")) +
    ggplot2::scale_fill_manual(name = NULL, values = c('#e41a1c')) +
    ggplot2::scale_color_manual(values=c('#e41a1c'))

gg_ca_pmf = ca_pmf %>%
    dplyr::filter(distance < 6) %>%
    dplyr::mutate(pmf = pmf - last(pmf)) %>% 
    ggplot2::ggplot(ggplot2::aes(x = distance, y = pmf)) +
    ggplot2::geom_smooth(ggplot2::aes(color='#377eb8'), size=2) + 
    ggplot2::geom_point(alpha=0.5) +
    ggplot2::xlab('Ion-lipid distance (Å)') + 
    ggplot2::ylab('') + 
    ggplot2::ggtitle(expression(paste("Ion-lipid PMF (kcal/mol)"))) + 
    ggplot2::theme_bw() +
    ggplot2::scale_x_continuous(breaks=seq(0, 6, 1), limits=c(1, 6)) + 
    ggplot2::theme(legend.position="none",  plot.margin = ggplot2::unit(c(0,0,0,0), "cm")) +
    ggplot2::scale_fill_manual(name = NULL, values = c('#377eb8')) +
    ggplot2::scale_color_manual(values=c('#377eb8'))

gg_mg_pmf = mg_pmf %>%
    dplyr::filter(distance < 6) %>%
    dplyr::mutate(pmf = pmf - last(pmf)) %>% 
    ggplot2::ggplot(ggplot2::aes(x = distance, y = pmf)) +
    ggplot2::geom_smooth(ggplot2::aes(color='#e41a1c'), size=2) + 
    ggplot2::geom_point(alpha=0.5) + 
    ggplot2::xlab('Ion-lipid distance (Å)') + 
    ggplot2::ylab('') + 
    ggplot2::ggtitle(expression(paste("Ion-lipid PMF (kcal/mol)"))) + 
    ggplot2::theme_bw() +
    ggplot2::scale_x_continuous(breaks=seq(0, 6, 1), limits=c(1, 6)) + 
    ggplot2::theme(legend.position="none", plot.margin = ggplot2::unit(c(0,0,0,0), "cm")) +
    ggplot2::scale_fill_manual(name = NULL, values = c('#e41a1c')) +
    ggplot2::scale_color_manual(values=c('#e41a1c'))


gg = cowplot::plot_grid(gg_ca, gg_mg, gg_ca_pmf, gg_mg_pmf, ncol=2, nrow=2, align='hv')
gg
```
